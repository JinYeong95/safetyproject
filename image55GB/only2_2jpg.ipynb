```python
import os
import shutil

# 텍스트 파일이 저장된 폴더 경로와 이미지 파일이 저장될 폴더 경로를 지정합니다.
text_folder_path = "C:/activatedproject/image55GB/only2label2"
image_folder_path = "C:/activatedproject/image55GB/output_img2"
output_image_folder_path = "C:/activatedproject/image55GB/only2_outputimg2"

# 출력 폴더가 없으면 생성합니다.
os.makedirs(output_image_folder_path, exist_ok=True)

# 텍스트 폴더 내의 파일들을 검색하고 조건에 맞게 이미지 파일을 분리합니다.
for file_name in os.listdir(text_folder_path):
    if file_name.endswith(".txt"):
        text_file_path = os.path.join(text_folder_path, file_name)
        image_name = os.path.splitext(file_name)[0]  # 확장자를 제외한 이미지 이름을 추출합니다.
        
        # 이미지 파일을 찾아서 해당 폴더로 옮깁니다.
        image_extensions = [".jpg", ".png"]
        for ext in image_extensions:
            source_image_path = os.path.join(image_folder_path, image_name + ext)
            if os.path.exists(source_image_path):
                output_image_path = os.path.join(output_image_folder_path, image_name + ext)
                shutil.move(source_image_path, output_image_path)
                print(f"이미지 파일 {image_name + ext}을(를) 옮겼습니다.")

print("작업이 완료되었습니다.")
```
