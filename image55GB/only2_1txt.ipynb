```python
import os
import shutil

# 입력 폴더와 출력 폴더의 경로를 지정합니다.
input_folder_path = "C:/activatedproject/image55GB/labels1"
output_folder_path = "C:/activatedproject/image55GB/only2label1"

# 출력 폴더가 없으면 생성합니다.
os.makedirs(output_folder_path, exist_ok=True)

# 입력 폴더 내의 파일들을 검색하고 조건에 맞게 파일을 분리합니다.
for file_name in os.listdir(input_folder_path):
    if file_name.endswith(".txt"):
        input_file_path = os.path.join(input_folder_path, file_name)
        
        with open(input_file_path, "r") as input_file:
            first_line = input_file.readline().strip()  # 파일의 첫 번째 줄을 읽습니다.
            first_number = int(first_line.split()[0]) if first_line else None  # 첫 번째 숫자를 추출합니다.

        if first_number == 2:
            output_file_path = os.path.join(output_folder_path, file_name)
            shutil.move(input_file_path, output_file_path)
            print(f"파일 {file_name}을(를) 옮겼습니다.")

print("작업이 완료되었습니다.")
```
